```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_bank = pd.read_csv('bank-additional-full.csv',delimiter=';')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_bank['job'] = le.fit_transform(df_bank['job'])
df_bank['marital'] = le.fit_transform(df_bank['marital'])
df_bank['education'] = le.fit_transform(df_bank['education'])
df_bank['default'] = le.fit_transform(df_bank['default'])
df_bank['housing'] = le.fit_transform(df_bank['housing'])
df_bank['loan'] = le.fit_transform(df_bank['loan'])
df_bank['contact'] = le.fit_transform(df_bank['contact'])
df_bank['month'] = le.fit_transform(df_bank['month'])
df_bank['poutcome'] = le.fit_transform(df_bank['poutcome'])
df_bank['day_of_week'] = le.fit_transform(df_bank['day_of_week'])
df_bank['y'] = le.fit_transform(df_bank['y'])
```

```python
df_bank.head()
```

```python
df_bank.describe().T
```

```python
df_bank.y.value_counts()
```

```python
from sklearn.utils import resample
```

```python
df_majority = df_bank[df_bank.y == 0]
df_minority = df_bank[df_bank.y == 1]
```

```python
df_sample = resample(df_minority,replace=True,n_samples=36548)
```

```python
df_sample = pd.concat([df_majority,df_sample])
```

```python
df_sample.y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
X = df_sample.drop('y',axis = 1)
Y = df_sample['y']
```

```python
Xtrain,Xval,ytrain,yval = train_test_split(X,Y,test_size = 0.3,random_state= 1)
```

```python
print(Xtrain.shape)
print(ytrain.shape)
print(Xval.shape)
print(yval.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(Xtrain,ytrain)
```

```python
model.score(Xtrain,ytrain)
```

```python
y_predict = model.predict(Xval)
```

```python
accuracy_score(yval,y_predict)
```

```python
confusion_matrix(yval,y_predict)
```

```python
def generate_models(model):
    model.fit(Xtrain,ytrain)
    print(model.score(Xtrain,ytrain))
    y_predict = model.predict(Xval)
    print(accuracy_score(yval,y_predict))
    print(confusion_matrix(yval,y_predict))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```
